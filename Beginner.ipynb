# --> Feature Engineering
Feature engineering is the process of using domain knowledge to extract features from raw data. These features can be used to improve the performance of machine learning algorithms.

![Data Science workflow](images/ml_workflow.png)


## 1. Feature Transformation

**Definition:**  
Feature transformation refers to preparing and modifying raw data to make it more suitable for modeling. Think of it as refining ingredients before cooking — better preparation leads to better results and more accurate machine learning models.



### 🔹  Missing Value Imputation  

**Definition:**  
Filling in or handling missing data values to prevent errors during training and avoid biased results.  
Common methods include:
- Mean/Median substitution  
- Forward/Backward fill  
- Predictive model-based imputation  

<img src="images/missing.png" alt="Missing Values" width="400"/>


### 🔹 Handling Categorical Features  

**Definition:**  
Categorical features are non-numeric data (such as labels, brands, or colors). Since machine learning algorithms work with numbers, these features must be converted into a numerical format.


#### Common Encoding Techniques:

- **Ordinal Encoding (for Input Features with Order):**  
  Assigns integer values to categories based on order.  
  Suitable when the categories have a **natural ranking** (e.g., Low < Medium < High).  
  Example:  
  `{'Low': 1, 'Medium': 2, 'High': 3}`  

- **Label Encoding (for Target/Output Variable):**  
  Assigns a unique integer to each class label.  
  Commonly used to encode the **target/output (dependent) variable** in classification problems.  
  Example:  
  `{'Cat': 0, 'Dog': 1, 'Rabbit': 2}`  

- **One-Hot Encoding (for Input Features without Order):**  
  Creates separate binary columns for each category (1 = present, 0 = not).  
  Prevents algorithms from assuming any ordinal relationship between categories.  
  Best for **nominal input features** with limited unique values.  
  Example:

  | Color | Red | Green | Blue |
  |-------|-----|-------|------|
  | Red   |  1  |   0   |  0   |
  | Green |  0  |   1   |  0   |
  | Blue  |  0  |   0   |  1   |

<img src="images/encode.png" alt="Encoding" width="400"/>



### 🔹  Outlier Detection  

**Definition:**  
Identifying and managing extreme values or unusual observations that can distort model training.  
Detection methods:
- Statistical: IQR, Z-score  
- Visual: Boxplots, Scatter plots  

Treatment options:
- Removal  
- Capping  
- Transformation  

<img src="images/outlier.png" alt="Outliers" width="400"/>



### 🔹 Feature Scaling

**Definition:**  
Feature scaling refers to **normalizing or standardizing** numerical values so that they contribute equally during model training.  
It's especially important for algorithms that rely on **distance metrics** (e.g., *KNN*, *SVM*) or **gradient descent**.


### Common Feature Scaling Techniques

1. **Standardization (Z-score Normalization)**  
   Transforms data to have a **mean of 0** and **standard deviation of 1**:

   $$
   z = \frac{x - \mu}{\sigma}
   $$

2. **Min-Max Normalization**  
   Scales features to a fixed range, typically **[0, 1]**:

   $$
   x_{\text{norm}} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
   $$

3. **Robust Scaling**  
   Scales using the **median** and **IQR** (Interquartile Range), ideal for data with **outliers**:

   $$
   x_{\text{scaled}} = \frac{x - \text{median}}{\text{IQR}}
   $$

4. **Mean Normalization**  
   Centers data to zero and scales by the **range**:

   $$
   x_{\text{mean\_norm}} = \frac{x - \mu}{x_{\max} - x_{\min}}
   $$

5. **Max Absolute Scaling**  
   Scales data by dividing by the **maximum absolute value**, keeping values in **[-1, 1]**:

   $$
   x_{\text{scaled}} = \frac{x}{\max(|x|)}
   $$


> 📌 Choose the scaler based on your data:
> - Use **StandardScaler** if features are normally distributed.
> - Use **RobustScaler** if data contains outliers.
> - Use **MinMaxScaler** or **MaxAbsScaler** for neural networks or sparse data.
---

### 🔹 Mathematical Transformations in ML (Overview)

Used to normalize distributions, reduce skewness, and improve model performance.



####  Function Transformations
Apply direct math functions:
- `log(x+1)` – right-skewed data
- `1/(x+ε)` – long tail
- `x²`, `√x` – skew adjustment
- `exp(x)` – left-skewed data



####  Power Transformations
Make data Gaussian-like:
- **Box-Cox** – positive values only
- **Yeo-Johnson** – handles negatives too

```python
PowerTransformer(method='yeo-johnson')
```



####  Quantile Transformation
Maps data to uniform/normal distribution using quantiles.
- Great for outliers
- May distort relationships

```python
QuantileTransformer(output_distribution='normal')
```



> ✅ Tip: Use visualizations to compare before/after transformations.



## 2. Feature Construction

**Definition:**  
Feature construction is the process of creating new features from existing data to improve the model’s performance by giving it more meaningful information.

**Example:**
- From `Height` and `Weight`, create a new feature `BMI = Weight / Height²`
- From `Date of Birth`, create a new feature `Age`

```python
df['BMI'] = df['Weight_kg'] / (df['Height_m'] ** 2)
df['Age'] = 2025 - pd.DatetimeIndex(df['DOB']).year
```

---

## 3. Feature Selection

**Definition:**  
Feature selection is the process of selecting the most relevant features and removing less useful ones to improve model accuracy and reduce overfitting.

**Techniques:**
- **Filter methods** (e.g., correlation, chi-square)
- **Wrapper methods** (e.g., RFE)
- **Embedded methods** (e.g., Lasso, Tree-based)

**Example:**
Select top 5 features using ANOVA F-value:

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=5)
X_selected = selector.fit_transform(X, y)
```

---

## 4. Feature Extraction

**Definition:**  
Feature extraction transforms original features into a new feature set (often with fewer dimensions), while preserving as much information as possible.

**🛠️ Techniques:**
- PCA (Principal Component Analysis)
- TF-IDF or Word2Vec (for text)
- Autoencoders (for deep learning)

**Example:**
Reduce dimensions from 100 to 10 using PCA:

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)
X_pca = pca.fit_transform(X)
```

---

### Comparison Table

| Technique              | Purpose                                         | Examples                            |
|------------------------|-------------------------------------------------|-------------------------------------|
| Feature Transformation | Modify existing features for better learning    | Scaling, Encoding, Log Transform    |
| Feature Construction   | Create new features from existing ones          | BMI, Age from DOB, Total Spend      |
| Feature Selection      | Keep only the most relevant features            | SelectKBest, Lasso, Recursive Elim  |
| Feature Extraction     | Transform data to a new feature space           | PCA, TF-IDF, Autoencoders           |

#### Why Scaling is Important

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize models
knn = KNeighborsClassifier()
knn_scaled = KNeighborsClassifier()

# Train on unscaled data
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Train on scaled data
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)

# Evaluate
print("Accuracy (Unscaled):", accuracy_score(y_test, y_pred))
print("Accuracy (Scaled):", accuracy_score(y_test, y_pred_scaled))
````
